# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y=[2,3,9,12]
```

```python
plt.plot(x)
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 11], [2,4,5,3]);
```

```python
plt.plot(x, y, 'r--')
plt.axis([0, 5, 0, 14])
plt.show()
```

```python
t = np.array(x+y)
# red circles, blue squares and green triangles
plt.plot(t, t, 'ro', t, t**2, 'bs', t, t*3, 'g^')
plt.show()
```

```python
plt.plot(x, y, linewidth=5.0);
plt.show()
```

```python
line, = plt.plot(x, y)
line.set_antialiased(False) # turn off antialising
plt.show()
```

```python
line, = plt.plot(x, y)
line.set_antialiased(True) # turn off antialising
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'r')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'k')
plt.show()
```
